```python
# normalizing data 
from sklearn import preprocessing
df2 = df
df2 = df2.drop(columns='name')
df2 = df2.drop(columns='artists')
df2 = df2.drop(columns='id')
df2 = df2.drop(columns='release_date')
x = df2.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df2 = pd.DataFrame(x_scaled,columns=df2.columns)
print(df2)
```

```python
# calculating varience of each numeric feature after normalizing
# song duration and loudness are found to have very low varience
df2.var()
```

```python
# generating box plots for numeric variables to check for univariate outliers
ax = df.boxplot(column=['energy'])
ax.set_title("box plots")
```

```python
# generating lower quartile, upper quartile and inter quartile range for each numeric variable in dataset.
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
# for numeric outlier removal interquartile range multiplier(k) is set at 1.5
dfx = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
dfx.shape
```

```python
# dfx is our training data frame
# we remove unnecessary features
# year and loudness are removed because they are highly correlated with other features
# duration is removed because it has very low varience
# artist and song name are unnecessary features
index = dfx['id']
dfx = dfx.drop(columns=['artists','name','release_date','year','loudness','duration_ms','id'],axis=1)
```

```python
# normalized data for KNN so that every feature plays an equal role
x = dfx.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
dfx = pd.DataFrame(x_scaled,columns=dfx.columns)
dfx.head()
```

```python
# importing KNN 
from sklearn.neighbors import NearestNeighbors
from scipy.sparse import csr_matrix
dfx.index = index
dfx.head()
```

```python
# k choosed to be 20 
# model_knn = NearestNeighbors(algorithm='kd_tree',n_neighbors=20)
import numpy as np

 #weighted KNN
weights = np.random.choice(dfx.shape[1])
model_knn = NearestNeighbors(algorithm='brute',metric='wminkowski',metric_params={'w': weights}, p=2, n_jobs=-1)


mat_songs = csr_matrix(dfx.values)
```

```python
# model training
model_knn.fit(dfx.values)
```

```python
# recommending songs
def recommend(idx, model, number_of_recommendations=5):
    query = dfx.loc[idx].to_numpy().reshape(1,-1)
    print('Searching for recommendations...')
    distances, indices = model.kneighbors(query,n_neighbors = number_of_recommendations)
    print(distances+indices)
    
    
    for i in indices:
        print(df[['name','artists']].loc[i].where(df['id']!=idx).dropna())
```

```python
#Tester

name = input('Enter song title: ')
print('Search results: ')
print(df[['artists','name']].where(df['name'] == name).dropna())

ind = int(input('Enter the index value of the required song: '))
idx = df['id'].loc[ind]

song = df['name'].loc[ind]
artists = df['artists'].loc[ind]

print('Song selected is ', song, 'by', artists)

nor = int(input('Enter number of recommendations: '))

recommend(idx, model_knn, nor)
```
